```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
dados.head(10)
```

# Reference

https://pandas.pydata.org/pandas-docs/version/0.22/api.html#dataframe

```python
dados['Valor'].mean()
```

# Separando por bairro

```python
dados['Bairro'].unique()
```

```python
bairros = ['Copacabana', 'Centro', 'Higienópolis', 'Cachambi', 
         'Barra da Tijuca', 'Ramos', 'Grajaú', 'Lins de Vasconcelos', 'Taquara']
bairros
```

```python
selecao = dados['Bairro'].isin(bairros)
dados = dados[selecao]
dados.head(10)
```

```python
dados['Bairro'].drop_duplicates()
```

# Realizando o Agrupamento

```python
grupo_bairro = dados.groupby('Bairro')
type(grupo_bairro)
```

```python
grupo_bairro.groups
```

# Calculando o valor médio de cada bairro

## Forma 1

```python
for bairro, data in grupo_bairro:
    print('{} -> {}'.format(bairro, data['Valor'].mean()))
```

## Forma 2 - Valor médio do bairro e também o valor do condominio

```python
grupo_bairro['Valor', 'Condominio'].mean().round(2)
```

# Estatistica Descritivvas

```python
grupo_bairro['Valor'].describe().round(2)
```

```python
grupo_bairro['Valor'].aggregate(['min', 'max', 'sum']).rename(columns = {'min': 'Mínimo', 'max': 'Máximo', 'sum': 'Somatório'})
```

# Análise gráfica

```python
dados2 = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
grupo_bairro2 = dados2.groupby('Bairro')
```

```python
# Gera figura estática para o rodar no jupyter
%matplotlib inline 

# Importando
import matplotlib.pyplot as plt # 
plt.rc('figure', figsize = (20, 10))
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color = 'blue') # plota a media com a cor azul
fig.set_ylabel('Valor do Aluguel') # Nome do eixo
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```

```python
fig2 = grupo_bairro2['Valor'].mean().plot.bar(color = 'blue') # plota a media com a cor azul
fig2.set_ylabel('Valor do Aluguel') # Nome do eixo
fig2.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```
