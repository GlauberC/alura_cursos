```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

# Verificar dados faltantes

```python
dados.isnull().head(10) #ou notnull
```

# Quantos dados são nulos ? 

```python
dados.info()
```

```python
dados[dados['Valor'].isnull()]
```

# Eliminando dados nulos baseado na coluna valor

```python
A = dados.shape[0]
dados.dropna(subset=['Valor'], inplace = True)
B = dados.shape[0]
A - B # Quantos foram eliminados
```

```python
dados[dados['Valor'].isnull()]
```

# Tratando dados de tipo apartamento que tenha condominio nulo

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

```python
A = dados.shape[0]
dados = dados[~selecao] # inverte a seleção
B = dados.shape[0]
A - B
```

# Atribuir valor 0  aos dados restantes

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
dados.head(10)
```

```python
dados = dados.fillna({'Condominio' : 0 , 'IPTU' : 0})
dados.head(10)
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
dados[dados['IPTU'].isnull()].shape[0]
```

```python
dados.info()
```

# Exportar novo arquivo

```python
dados.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
```
