# Ações

- Criar um dataframe referente a uma escola contendo id, nome, matricula, sexo, idade,  3 notas, media,recuperacao,media final e se está aprovado 
- importar o dataframe
- criar lista com ordem alfabética sem a coluna id
- exportar a lista tratada
- ordenar pela media 
- criar listas das 10 melhores notas de cada trimestre
- criar lista com as 5 meninas com as melhores medias e depois criar outra para os meninos
- criar 2 listas para aprovado e reprovado respectivamente
- criar lista todas as idades dos alunos na escola e quantos alunos de cada idade


```python
import pandas as pd
```

# importar o dataframe

```python
dados = pd.read_csv('dados/alunos.csv', sep=';')
dados
```

# criar lista com ordem alfabética sem a coluna id

```python
dados_semId = dados.iloc[:, 1:]
dados_ordenados = dados_semId.sort_values(by='Nome')
dados_ordenados.index = range(dados_semId.shape[0])
dados_ordenados
```

# exportar a lista tratada

```python
dados_ordenados.to_csv('dados/dados_tratados.csv', sep=';', index = False)
```

# ordenar pela media 

```python
dados = dados_ordenados
dados_ordenados_media = dados.sort_values(by="Média Final", ascending=False)
dados_ordenados_media
```

# criar listas das 10 melhores notas de cada trimestre

```python
melhores_tri1 = dados.sort_values(by="Nota 1", ascending=False).head(10)
melhores_tri1
```

```python
melhores_tri2 = dados.sort_values(by="Nota 2", ascending=False).head(10)
melhores_tri2
```

```python
melhores_tri3 = dados.sort_values(by="Nota 3", ascending=False).head(10)
melhores_tri3
```

# criar lista com as 5 meninas com as melhores medias e depois criar outra para os meninos

```python
selecao = dados['Sexo'] == 'M'
melhores_meninos = dados[selecao].sort_values(by = "Média Final", ascending = False).head()
melhores_meninos
```

```python
selecao = dados['Sexo'] == 'F'
melhores_meninas = dados[selecao].sort_values(by = 'Média Final', ascending = False).head()
melhores_meninas
```

# criar 2 listas para aprovado e reprovado respectivamente

```python
selecao = dados['Situação'] == 'APROVADO'
lista_aprovados = dados[selecao]
lista_aprovados
```

```python
selecao = dados['Situação'] == 'REPROVADO'
lista_reprovados = dados[selecao]
lista_reprovados
```

# criar lista todas as idades dos alunos na escola e quantos alunos de cada idade

```python
idades = dados['Idade'].sort_values().drop_duplicates()
idades
```
