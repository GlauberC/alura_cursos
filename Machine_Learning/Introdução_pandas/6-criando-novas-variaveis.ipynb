```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

# Criando variável valor bruto

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head(10)
```

# Criando variável valor por m²

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
dados.head(10)
```

# Arredondando as casas decimais

```python
dados['Valor m2'] = dados['Valor m2'].round(2)
dados.head(10)
```

# Criando variável valor bruto por m²

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
dados.head(10)
```

# Criando variável a partir do valor do Tipo

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados.head(50)
```

# Excluindo variáveis

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
dados_aux.head()
```

## Método 1

```python
del dados_aux['Valor Bruto']
dados_aux.head()
```

## Método 2

```python
dados_aux.pop('Valor Bruto m2')
dados_aux.head()
```

## Método 3

```python
dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace = True) # axis = 1 para informar que a remoção é nas colunas (linhas -> axis = 0)
dados.head()
```

# Exportando dados

```python
dados.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
```
